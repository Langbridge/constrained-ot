```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='latin1')
    return dict
```

```python
datadict = unpickle('/Users/abilangbridge/Downloads/cifar-10-batches-py/test_batch')

X = datadict["data"] 
Y = datadict['labels']
X = X.reshape(10000, 3, 32, 32).transpose(0,2,3,1).astype("uint8")
Y = np.array(Y)
```

```python
# labels:
# 0: plane
# 1: car
# 2: bird
# 3: cat
# 4: deer
# 5: dog
# 6: frog
# 7: horse
# 8: ship
# 9: truck
```

```python
subset_idxs = np.where(Y < 4)[0]

fig, axes1 = plt.subplots(5,5,figsize=(6,6))
for j in range(5):
    for k in range(5):
        i = np.random.choice(subset_idxs)
        axes1[j][k].set_axis_off()
        axes1[j][k].imshow(X[i:i+1][0], interpolation='sinc')
        axes1[j][k].set_title(Y[i])

fig.tight_layout()
```
